```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import joblib
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

n_steps = 60
epochs = 50
batch_size = 32

df = pd.read_csv('../../data/cleaned/crypto_dataset_cleaned.csv', parse_dates=['Date'], index_col='Date')
rep_df = pd.read_csv('../../data/processed/representative_coins.csv')
rep_coins = rep_df['Representative_Coin'].tolist()

output_dir = '../../results/LSTM_csv/separate_coins'
os.makedirs(output_dir, exist_ok=True)
metrics_records = []

def generate_signals(prices):
    signals = ['Hold']
    for i in range(1, len(prices)):
        if prices[i] > prices[i-1]:
            signals.append('Buy')
        elif prices[i] < prices[i-1]:
            signals.append('Sell')
        else:
            signals.append('Hold')
    return signals

def create_sequences(data, n_steps):
    X, y = [], []
    for i in range(len(data) - n_steps):
        X.append(data[i:i+n_steps])
        y.append(data[i+n_steps])
    return np.array(X), np.array(y)

for coin in rep_coins:
    print(f"\n= LSTM Modeling for {coin} =")

    prices = df[[coin]].values.astype('float64')
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_prices = scaler.fit_transform(prices)

    X, y = create_sequences(scaled_prices, n_steps)
    X = X.reshape((X.shape[0], X.shape[1], 1))

    train_size = int(len(X) * 0.8)
    X_train, X_test = X[:train_size], X[train_size:]
    y_train, y_test = y[:train_size], y[train_size:]

    model = Sequential([
        LSTM(50, return_sequences=True, input_shape=(n_steps, 1)),
        Dropout(0.2),
        LSTM(50),
        Dropout(0.2),
        Dense(1)
    ])

    model.compile(optimizer='adam', loss='mean_squared_error')
    es = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

    model.fit(
        X_train, y_train,
        validation_data=(X_test, y_test),
        epochs=epochs,
        batch_size=batch_size,
        callbacks=[es],
        shuffle=False,
        verbose=1
    )

    # Predict on entire dataset sequences (train + test)
    y_pred_scaled_full = model.predict(X)
    y_pred_full = scaler.inverse_transform(y_pred_scaled_full)

    all_dates = df.index[n_steps:]
    y_actual_full = df.loc[all_dates, coin].values

    # Compute metrics only on test portion
    y_actual_test = y_actual_full[train_size:]
    y_pred_test = y_pred_full[train_size:]

    rmse = np.sqrt(mean_squared_error(y_actual_test, y_pred_test))
    mae = mean_absolute_error(y_actual_test, y_pred_test)
    mape = np.mean(np.abs((y_actual_test - y_pred_test) / (y_actual_test + 1e-10))) * 100
    r2 = r2_score(y_actual_test, y_pred_test)

    print(f"\n#-------------- LSTM Evaluation Results ({coin}) --------------#")
    print(f"| {'Metric':<8} | {'Value':>12} |")
    print(f"|{'-'*10}|{'-'*14}|")
    print(f"| RMSE    | {rmse:>12.4f} |")
    print(f"| MAE     | {mae:>12.4f} |")
    print(f"| MAPE    | {mape:>12.2f} |")
    print(f"| R2      | {r2:>12.4f} |")
    print("#-------------------------------------------------------------#")

    signals_full = generate_signals(y_pred_full.flatten())

    hist_forecast_df = pd.DataFrame({
        'Date': all_dates,
        'Actual': y_actual_full,
        'Predicted': y_pred_full.flatten(),
        'Signal': signals_full
    })

    hist_csv_path = os.path.join(output_dir, f"{coin}_historical_forecast.csv")
    hist_forecast_df.to_csv(hist_csv_path, index=False)

    signals_test = generate_signals(y_pred_test.flatten())
    test_forecast_df = pd.DataFrame({
        'Date': all_dates[train_size:],
        'Actual': y_actual_test,
        'Predicted': y_pred_test.flatten(),
        'Signal': signals_test
    })
    test_csv_path = os.path.join(output_dir, f"{coin}_test_forecast.csv")
    test_forecast_df.to_csv(test_csv_path, index=False)

    future_dates = pd.date_range(start=df.index[-1] + pd.Timedelta(days=1), periods=30, freq='D')
    last_seq = scaled_prices[-n_steps:].reshape(1, n_steps, 1)
    future_preds_scaled = []
    for _ in range(30):
        next_pred = model.predict(last_seq)[0, 0]
        future_preds_scaled.append(next_pred)
        last_seq = np.append(last_seq[:, 1:, :], [[[next_pred]]], axis=1)
    future_preds_scaled = np.array(future_preds_scaled).reshape(-1, 1)
    future_pred = scaler.inverse_transform(future_preds_scaled)
    signals_future = generate_signals(future_pred.flatten())

    future_forecast_df = pd.DataFrame({
        'Date': future_dates,
        'Predicted': future_pred.flatten(),
        'Signal': signals_future
    })
    future_csv_path = os.path.join(output_dir, f"{coin}_future_forecast.csv")
    future_forecast_df.to_csv(future_csv_path, index=False)

    print(f"Saved historical forecast CSV: {hist_csv_path}")
    print(f"Saved test forecast CSV: {test_csv_path}")
    print(f"Saved future forecast CSV: {future_csv_path}")

    os.makedirs('../../saved_models', exist_ok=True)
    model_path = f"../../saved_models/{coin}_lstm_model.h5"
    scaler_path = f"../../saved_models/{coin}_lstm_scaler.pkl"
    model.save(model_path)
    joblib.dump(scaler, scaler_path)
    print(f"Saved model and scaler for {coin}")

    metrics_records.append({'Coin': coin, 'RMSE': rmse, 'MAE': mae, 'MAPE': mape, 'R2': r2})

    # Plotting
    plt.figure(figsize=(14,6))
    plt.plot(df.index, df[coin].values, label='Actual Price', color='blue')
    plt.plot(all_dates, y_pred_full.flatten(), label='Predicted Price (Train+Test)', color='red', linestyle='--')
    plt.plot(future_dates, future_pred, label='30-Day Future Forecast', color='green', linestyle='-.')
    plt.fill_between(future_dates, future_pred.flatten(), alpha=0.2, color='green')
    plt.title(f"LSTM Prediction and 30-Day Forecast for {coin}", fontsize=16)
    plt.xlabel("Date")
    plt.ylabel("Price (USD)")
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

metrics_df = pd.DataFrame(metrics_records)
metrics_summary_path = os.path.join(output_dir, "lstm_metrics_summary.csv")
metrics_df.to_csv(metrics_summary_path, index=False)
print(f"Saved combined LSTM metrics CSV: {metrics_summary_path}")
```
