```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from pmdarima import auto_arima
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Load cleaned data and representative coins
df = pd.read_csv('../../data/cleaned/crypto_dataset_cleaned.csv', parse_dates=['Date'], index_col='Date')
rep_df = pd.read_csv('../../data/processed/representative_coins.csv')
rep_coins = rep_df['Representative_Coin'].tolist()

def test_stationarity(series):
    result = adfuller(series.dropna())
    return result[1] <= 0.05

def generate_signals(prices):
    signals = ['Hold']
    for i in range(1, len(prices)):
        if prices[i] > prices[i-1]:
            signals.append('Buy')
        elif prices[i] < prices[i-1]:
            signals.append('Sell')
        else:
            signals.append('Hold')
    return signals

# Output directory for CSVs
output_dir = '../../results/Arima_csv/separate_coins'
os.makedirs(output_dir, exist_ok=True)

metrics_list = []

for coin in rep_coins:
    print(f"\nProcessing {coin}")

    series = df[coin].copy().ffill()
    log_series = np.log(series)

    diff_order = 0
    tmp_series = log_series.copy()
    while not test_stationarity(tmp_series) and diff_order < 2:
        diff_order += 1
        tmp_series = tmp_series.diff().dropna()
    stationary_series = tmp_series

    train_size = int(len(stationary_series) * 0.8)
    train, test = stationary_series.iloc[:train_size], stationary_series.iloc[train_size:]

    model = auto_arima(
        train,
        start_p=1, max_p=5,
        start_q=1, max_q=5,
        d=0,
        seasonal=False,
        stepwise=True,
        trace=True,
        error_action='ignore',
        suppress_warnings=True,
        random_state=42
    )

    if model.order == (0, 0, 0):
        manual_model = ARIMA(train, order=(3, 0, 3)).fit()
        forecast_log_values = manual_model.forecast(steps=len(test))
        fitted_log = manual_model.predict(start=0, end=len(stationary_series) - 1)
    else:
        forecast_log_values, _ = model.predict(n_periods=len(test), return_conf_int=True)
        fitted_log = model.predict_in_sample()

    # Convert fitted (in-sample prediction) log to prices
    first_log_val = log_series.iloc[0]
    fitted_cumsum = fitted_log.cumsum()
    fitted_prices = np.exp(fitted_cumsum + first_log_val)

    # Align actual prices length with fitted prices length properly for BETH-USD and keep others intact
    # Get corresponding start index in original series for stationary series
    start_idx = series.index.get_loc(stationary_series.index[0])
    actual_for_fitted = series.iloc[start_idx : start_idx + len(fitted_prices)]

    # Compute and print metrics on in-sample prediction
    rmse = np.sqrt(mean_squared_error(actual_for_fitted.values, fitted_prices.values))
    mae = mean_absolute_error(actual_for_fitted.values, fitted_prices.values)
    mape = np.mean(np.abs((actual_for_fitted.values - fitted_prices.values) / (actual_for_fitted.values + 1e-10))) * 100
    r2 = r2_score(actual_for_fitted.values, fitted_prices.values)
    print(f"Metrics on in-sample prediction for {coin}: RMSE={rmse:.4f}, MAE={mae:.4f}, MAPE={mape:.2f}%, R2={r2:.4f}")

    metrics_list.append({
        'Coin': coin,
        'RMSE': rmse,
        'MAE': mae,
        'MAPE': mape,
        'R2': r2
    })

    # Convert forecast log values to prices for out-of-sample test period
    last_train_log = log_series.iloc[train_size + diff_order - 1]
    forecast_log = np.concatenate([[last_train_log], forecast_log_values])
    for i in range(1, len(forecast_log)):
        forecast_log[i] += forecast_log[i-1]
    forecasted_prices = np.exp(forecast_log[1:])
    test_actual_prices = series.iloc[train_size + diff_order : train_size + diff_order + len(test)]

    # Generate signals
    historical_signals = generate_signals(fitted_prices.values)
    forecast_signals = generate_signals(forecasted_prices)

    # Save CSV: in-sample prediction & actuals with signals (historical forecast)
    hist_forecast_df = pd.DataFrame({
        'Date': actual_for_fitted.index,
        'Actual': actual_for_fitted.values,
        'Forecast': fitted_prices.values,
        'Signal': historical_signals
    })
    hist_forecast_df.to_csv(os.path.join(output_dir, f'{coin}_historical_forecast.csv'), index=False)

    # Save CSV: out-of-sample forecast vs actual with signals (test period)
    test_forecast_df = pd.DataFrame({
        'Date': test_actual_prices.index,
        'Actual': test_actual_prices.values,
        'Forecast': forecasted_prices,
        'Signal': forecast_signals
    })
    test_forecast_df.to_csv(os.path.join(output_dir, f'{coin}_test_forecast.csv'), index=False)

    # Future predictions for 30 days beyond last date
    future_periods = 30
    last_date = series.index[-1]
    future_dates = pd.date_range(start=last_date + pd.Timedelta(days=1), periods=future_periods, freq='D')

    if model.order == (0, 0, 0):
        future_forecast_log = manual_model.forecast(steps=future_periods)
    else:
        future_forecast_log, _ = model.predict(n_periods=future_periods, return_conf_int=True)

    last_log_val = log_series.iloc[-1]
    future_forecast_log = np.concatenate([[last_log_val], future_forecast_log])
    for i in range(1, len(future_forecast_log)):
        future_forecast_log[i] += future_forecast_log[i-1]
    future_predicted_prices = np.exp(future_forecast_log[1:])
    future_signals = generate_signals(future_predicted_prices)

    future_forecast_df = pd.DataFrame({
        'Date': future_dates,
        'Predicted': future_predicted_prices,
        'Signal': future_signals
    })
    future_forecast_df.to_csv(os.path.join(output_dir, f'{coin}_future_forecast.csv'), index=False)

    # Plot all lines together
    plt.figure(figsize=(14, 7))
    plt.plot(series.index, series.values, label='Actual Price', color='blue')
    plt.plot(actual_for_fitted.index, fitted_prices.values, label='Forecast (In-sample Prediction)', color='orange')
    plt.plot(test_actual_prices.index, forecasted_prices, label='Forecast (Out-of-sample Period)', color='red', linestyle='--')
    plt.plot(future_dates, future_predicted_prices, label='Forecast (Future 30-day Prediction)', color='green', linestyle=':')
    plt.title(f'ARIMA: Actual and Forecast vs Prediction for {coin}')
    plt.xlabel('Date')
    plt.ylabel('Price (USD)')
    plt.legend()
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()

# Save combined metrics summary for all coins
metrics_df = pd.DataFrame(metrics_list)
metrics_save_path = os.path.join(output_dir, 'arima_metrics_summary.csv')
metrics_df.to_csv(metrics_save_path, index=False)
print(f"Saved ARIMA performance metrics summary at {metrics_save_path}")
```
